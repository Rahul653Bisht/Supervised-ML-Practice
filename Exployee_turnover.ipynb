```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score
```

```python
employee_data = pd.read_csv("employee_turnover.csv")
```

```python
employee_data.head()
employee_data.info()
employee_data.head()
```

```python
employee_data["Employee_Turnover"].nunique()
```

```python
X = employee_data.drop(columns=["Employee_Turnover"])
Y = employee_data["Employee_Turnover"]
```

```python
#Train Test Split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#checking the y train value distribution
Y_train[Y_train==0]#533
Y_train[Y_train==1]#547
```

```python
#training logistic regression model
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(Y_test,Y_pred)
print("accuracy: ",acc)
pre = precision_score(Y_test,Y_pred)
print("precision: ",pre)
```

```python
#L1 Regularization (Lasso)
lasso = LogisticRegression(penalty="l1",solver='liblinear',C=0.5)
lasso.fit(X_train,Y_train)
```

```python
#L2 Regularization (Ridge)
ridge = LogisticRegression(penalty="l2",C=1,max_iter=200)
ridge.fit(X_train,Y_train)
```

```python
#Model Evaluation
from sklearn.metrics import accuracy_score,classification_report
models = {'Baseline':model,'Lasso':lasso,'Ridge':ridge}
for name,mod in models.items():
    Y_pred= mod.predict(X_test)
    print(f"\n{name}")
    print("Accuracy:",accuracy_score(Y_test,Y_pred))
    print(classification_report(Y_test,Y_pred))
```
