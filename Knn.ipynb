```python
import pandas as pd
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
heart_data = pd.read_csv("heart.csv")
heart_data.head()
```

```python
X = heart_data.drop(columns=["target"])
Y = heart_data["target"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_classifier.fit(X_train_scaled,Y_train)
```

```python
Y_pred = knn_classifier.predict(X_test_scaled)
```

```python
print("recall score:",recall_score(Y_test,Y_pred))
print("accuracy score:",accuracy_score(Y_test,Y_pred))
print("precision score:",precision_score(Y_test,Y_pred))
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=7)
knn_classifier.fit(X_train_scaled,Y_train)
```

```python
# model is giving best results for K - 7 for this data set
Y_pred= knn_classifier.predict(X_test_scaled)
print("recall score:",recall_score(Y_test,Y_pred))
print("accuracy score:",accuracy_score(Y_test,Y_pred))
print("precision score:",precision_score(Y_test,Y_pred))
```

```python
#Cross Validation For  hyperparameter tuning using GrideSearchCV
from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()
param_grid = {"n_neighbors":[3,5,7,9]}
classifierCV = GridSearchCV(classifier,param_grid,cv=5)
```

```python
classifierCV.fit(X_train_scaled,Y_train)
Y_pred = classifierCV.predict(X_test_scaled)
```

```python
print("recall score:",recall_score(Y_test,Y_pred))
print("accuracy score:",accuracy_score(Y_test,Y_pred))
print("precision score:",precision_score(Y_test,Y_pred))
```

```python
#Pipeline
from sklearn.pipeline import Pipeline
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
pipeline = Pipeline([("scaler",StandardScaler()),("knn",KNeighborsClassifier())])
```

```python
param_grid = {"knn__n_neighbors":[3,5,7,9]}
```

```python
classifierCV = GridSearchCV(pipeline,param_grid,cv=5,scoring="recall")
```

```python
classifierCV.fit(X_train,Y_train)
```

```python
Y_pred = classifierCV.predict(X_test)
```

```python
print("recall score:",recall_score(Y_test,Y_pred))
print("accuracy score:",accuracy_score(Y_test,Y_pred))
print("precision score:",precision_score(Y_test,Y_pred))
```
