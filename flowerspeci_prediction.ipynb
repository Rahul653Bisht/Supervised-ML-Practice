```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
```

```python
flower_data = pd.read_csv("Iris.csv")
```

```python
flower_data["Species"] = flower_data["Species"].map({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2})
```

```python
flower_data.head()
```

```python
X = flower_data.drop(columns=["Id","Species"])
Y = flower_data["Species"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
pipeline = Pipeline([("scaler",StandardScaler()),("knn",KNeighborsClassifier())])
```

```python
param_grid = {"knn__n_neighbors":[3,5,7,9]}
```

```python
knn = GridSearchCV(pipeline,param_grid,cv=5,scoring="accuracy")
```

```python
knn.fit(X_train,Y_train)
```

```python
knn_pred = knn.predict(X_test)
```

```python
knn_accuracy = accuracy_score(Y_test,knn_pred)
```

```python
print("KNN Accuracy:",knn_accuracy)
```

```python
pipeline_lr = Pipeline([("scaler",StandardScaler()),("lr",LogisticRegression())])
```

```python
pipeline_lr.fit(X_train,Y_train)
```

```python
lr_pred = pipeline_lr.predict(X_test)
```

```python
print("LR accuracy score:",accuracy_score(Y_test,lr_pred))
```
